```python
# Step 1: Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Step 2: Upload image
from google.colab import files
uploaded = files.upload()

filename=list(uploaded.keys())[0]
image =cv2.imread(filename)
cv2_imshow(image)

gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

resized=cv2.resize(gray,(500,500))
cv2_imshow(resized)

a=(image,gray)
for i in a:
  resized=cv2.resize(i,(200,200))
  cv2_imshow(i)

annotated = image.copy()
cv2.rectangle(annotated,(600,300),(200,1200),(0,255,255),2)
cv2.putText(annotated,"Rectangle",(500,1200),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,255),2)
cv2_imshow(annotated)


cv2.imwrite('output_image.jpg',annotated)
print("image saved as output_image_jpg")

from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array

uploaded=files.upload()
filename  =list(uploaded.keys())[0]
image=cv2.imread(filename)
cv2_imshow(image)

img = load_img(filename,target_size=(200,200))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array,axis=0)


datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.5, 1.5]
)
i = 0
for batch in datagen.flow(img_array, batch_size=1):
  plt.figure(figsize=(2,2))
  plt.imshow(batch[0].astype('uint8'))
  plt.axis('off')
  i += 1
  if i > 5:
    break
```
